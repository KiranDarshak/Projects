```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import spacy
import re
```

```python
nlp=spacy.load("en")
```

```python
data=nlp(u"Hi there, My name is KiranDarshak.Currently pursuing Data Analytics at Langara College, Vancouver. KiranDarshak is an Economics Graduate. KiranDarshak is aspiring to be a DataScientist in Future. Proficient in Python and RStudio. Passionate about Data Analytics and Python,RStudio. Worked on latest topics including Machine-Learning, Algorithms, Classification, PrincipalComponentAnalysis. NaturalLanguageProcessing. Regular posts in LinkedIn under 100daysofcode. Apart from studies, worked on publicly available datasets from Kaggle. Vivid user of MachineLearning concepts and SQL.")
```

## Step 1 Sentence Segmentation

```python
for sentence in data.sents:
    print(sentence)
```

## Step 2 Tokenization

```python
for tokens in data:
    print(tokens.text)
```

## Step 3 Parts Of Speech

```python
for tokens in data:
    print(tokens.text,"\t\t",tokens.pos_,"\t\t",spacy.explain(tokens.pos_))
```

# Lemmatization

```python
for tokens in data:
    print(tokens.text,"\t\t",tokens.lemma,"\t\t",tokens.lemma_)
```

# Visualizing the entity recognizer

```python
from spacy import displacy
displacy.render(data, style='dep', jupyter=True, options={'distance': 110})
```

```python
from spacy import displacy
displacy.render(data, style='ent', jupyter=True, options={'distance': 110})
```

```python
import matplotlib.pyplot as plt
#import WordCloud,STOPWORDS from wordcloud
from wordcloud import WordCloud, STOPWORDS

#Setting stopwords in a variable
stopwords = set(STOPWORDS)  

#creating function show_wordcloud
def show_wordcloud(insertdata, title = None):
    wordcloud = WordCloud(width = 720,
                          height=340,
        background_color='white',
        stopwords=stopwords,
        max_font_size=100, 
        scale=5,
    ).generate(str(insertdata))

    fig = plt.figure(1, figsize=(80,34))
    plt.axis('off')
    plt.imshow(wordcloud)
    plt.show()
show_wordcloud(data)
```
